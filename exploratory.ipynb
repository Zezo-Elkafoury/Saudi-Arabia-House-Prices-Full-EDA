### Outliers

```python
plt.figure(figsize=(8, 6))
sns.boxplot(data=df['size'])
plt.title('Box Plot of Duration (Outliers Shown as Dots)')
plt.show()
```

There are some dots which represent outliers

```python
def identify_outliers(df, column_names):
    for column_name in column_names:
        while True:
            Q1 = df[column_name].quantile(0.25)
            Q3 = df[column_name].quantile(0.75)
            IQR = Q3 - Q1
            lower_bound = Q1 - 1.5 * IQR
            upper_bound = Q3 + 1.5 * IQR
            initial_shape = df.shape[0]
            df = df[(df[column_name] >= lower_bound) & (df[column_name] <= upper_bound)]
            final_shape = df.shape[0]
            if initial_shape == final_shape:
                break
    return df

df = identify_outliers(df, ['size'])
```

##### Barplot function

```python
def plot_average_price(df, x_col, y_col, title, x_label, y_label, palette='viridis', figsize=(10, 6), fontsize=20):
    plt.figure(figsize=figsize)
    sns.barplot(x=x_col, y=y_col, data=df, palette=palette)

    # Format the plot with Arabic text
    plt.title(title, fontsize=fontsize, fontname='Arial')  # Arabic title
    plt.xlabel(x_label, fontsize=fontsize-2, fontname='Arial')  # Arabic x-axis label
    plt.ylabel(y_label, fontsize=fontsize-2, fontname='Arial')  # Arabic y-axis label
    plt.xticks(fontsize=fontsize-6)
    plt.yticks(fontsize=fontsize-6)
    plt.show()
```

### What are the most frequent words in the details column

```python

from collections import Counter
import re


# Combine all text in the 'details' column into a single string
text = ' '.join(df['details'].astype(str))  # Ensure all values are strings

# Reshape and apply bidi algorithm for proper Arabic rendering
# reshaped_text = arabic_reshaper.reshape(text)  # Reshape Arabic text
# bidi_text = get_display(reshaped_text)  # Apply bidi algorithm for RTL

# Extract words and count their frequency
words = re.findall(r'\b\w+\b', text) 
words=[x for x in words if x.lower() not in ['the','of','and','in','a','an','for','s','is']] # Tokenize the reshaped and RTL text
word_counts = Counter(words)

# Get the top 10 most common words
top_words = word_counts.most_common(15)

# Separate the words and their counts
words, counts = zip(*top_words)

# Plot the results
plt.figure(figsize=(10, 6))
sns.barplot(x=list(counts), y=list(words), palette='viridis')

# Format the plot with Arabic text
plt.title('Top 10 most frequent words in the details column', fontsize=20, fontname='Arial')  # Arabic title
plt.xlabel('Word', fontsize=18, fontname='Arial')  # Arabic x-axis label
plt.ylabel('Frequency', fontsize=18, fontname='Arial')  # Arabic y-axis label

# Ensure the y-axis labels (words) are displayed correctly
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

plt.show()
```

these words are possibly used to attract customers to the ads

### Average price of houses by each of the most frequent words in the descrption

```python
# Extract words from the 'details' column

# Calculate the average house price for each word
avg_prices = {}
for word, _ in top_words:
    filtered_df = df[df['details'].str.contains(word, case=False, na=False)]
    avg_prices[word] = filtered_df['price'].mean()

# Format the words in Arabic
formatted_words = [word for word in avg_prices.keys()]

# Plot the results
plt.figure(figsize=(14, 8))
plt.bar(formatted_words, avg_prices.values(), color='skyblue')
plt.xlabel('Words')
plt.ylabel('Average House Price')
plt.title('Average House Price for Top 10 Frequent Words in Details Column')
```

From this graph we can determine which words are used in marketing campaigns and which of them affect the price ofthe house
